## NetworkXの練習

演習1

```python
#必要ライブラリのインポート
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#無向グラフを作る
G = nx.Graph()
print(G)
```

```python
G.add_node(1) #ノード1を追加
print(G)
```

```python
G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(2, 3, weight=0.9) #エッジに重みを追加
print(G.number_of_nodes(), G.number_of_edges())
```

```python
G = nx.Graph()
edges = [(1,2), (2,3)]
G.add_edges_from(edges) #エッジのリストを追加する
print(G.nodes)
print(G.edges)
```

```python
G = nx.Graph()
edges = [(1,2,0.2), (2,3,0.9)] #重み付きエッジのリストを追加する
G.add_weighted_edges_from(edges)
nx.draw(G, with_labels=True)
```

```python
#有向グラフを作る
D = nx.DiGraph()
edges = [(1,2), (2,3)]
D.add_edges_from(edges)
nx.draw(D, with_labels=True)
```
